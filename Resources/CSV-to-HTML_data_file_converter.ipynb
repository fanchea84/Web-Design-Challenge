### Purpose of this Jupyter Notebook is to convert the PyWeather data from CSV to HTML format

```python
# Import Dependencies
import pandas as pd
from pprint import pprint
```

```python
# Make a reference to the books.csv file path
csv_path = "cities.csv"

# Import the cities.csv file as a DataFrame
cities_df = pd.read_csv(csv_path, encoding="utf-8")
cities_df.head()
```

```python
# Drop the UNNAMED:0 column from the DataFrame
cities_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
cities_df.head()
```

```python
# Convert DataFrame to HTML
awesomesauce = cities_df.to_html(index=False)
pprint(awesomesauce)
```
